```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pickle 
from os import path
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn import metrics
from sklearn import preprocessing
from sklearn.model_selection import train_test_split 
```

# Đọc dữ liệu và xử lý ban đâu

```python
data1=pd.read_csv("./data/Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv")
```

```python
# in ra tổng số lượng dữ liệu
print("Tổng số lượng dữ liệu:", data1.shape[0])

# in ra số lượng dữ liệu portscan
print("Số lượng dữ liệu portscan:", data1[data1[' Label'] == 'PortScan'].shape[0])

# in ra số lượng dữ liệu benign
print("Số lượng dữ liệu benign:", data1[data1[' Label'] == 'BENIGN'].shape[0])
```

```python
data2=pd.read_csv("./data/Monday-WorkingHours.pcap_ISCX.csv")
```

```python
# Lấy 300000 dữ liệu BEIGN của bộ dữ liệu Monday-WorkingHours.pcap_ISCX
benign_data = data2[data2[' Label'] == 'BENIGN'].head(300000)
```

```python
# thêm 300k dữ liệu BENIGN vào bộ dữ liệu  Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv
concatenate_data = pd.concat([benign_data, data1])
```

```python
#kiểm tra số lượng các loại của dữ liệu mới thêm
# in ra tổng số lượng dữ liệu
print("Tổng số lượng dữ liệu:", concatenate_data.shape[0])

# in ra số lượng dữ liệu portscan
print("Số lượng dữ liệu portscan:", concatenate_data[concatenate_data[' Label'] == 'PortScan'].shape[0])

# in ra số lượng dữ liệu benign
print("Số lượng dữ liệu benign:", concatenate_data[concatenate_data[' Label'] == 'BENIGN'].shape[0])
```

```python
#kiểm tra dữ liệu null 
print(concatenate_data.isnull().sum())
```

```python
# Xử lý dữ liệu infinity và NAN
concatenate_data = concatenate_data.replace([np.inf, -np.inf], 0)
```

```python
concatenate_data.fillna(0, inplace=True)
```

```python
print(type(concatenate_data))
```

```python
#Viết dữ liệu nối lại
concatenate_data.to_csv('Probe_Train.csv', index=False)
```

```python
#Tách cột Label ra khỏi dữ liệu
label = concatenate_data.pop(' Label') # tách cột Label khỏi bộ dữ liệu
```

```python
# Chuyển dữ liệu BENIGN thành 0, PortScan thành 1
label = label.replace('BENIGN', 0)
label = label.replace('PortScan', 1)
```

# Normalizing


```python
# Khởi tạo MinMaxScaler
scaler = MinMaxScaler()
```

```python
# Fit và transform dữ liệu
data = scaler.fit_transform(concatenate_data)
```

```python
concatenate_data = pd.DataFrame(data,columns=concatenate_data.columns)
```

```python
concatenate_data
```

# Split

```python
X_train, X_test = train_test_split(concatenate_data,test_size=0.30, random_state=50)
```

```python
Y_train, Y_test = train_test_split(label,test_size=0.30, random_state=50)
```
